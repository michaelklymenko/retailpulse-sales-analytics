```python
df[["Sales", "Quantity", "Discount", "Profit"]].describe().round(2)
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
```

```python
df.info()
```

```python
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
```

```python
df[["Order Date", "Ship Date"]]
```

```python
df["Year"] = df["Order Date"].dt.year
df["Month"] = df["Order Date"].dt.month
df["Month Name"] = df["Order Date"].dt.month_name()
```

```python
df["Order Year-Month"] = df["Order Date"].dt.to_period("M").astype(str)
```

```python
df["Profit Margin"] = df["Profit"] / df["Sales"]
```

```python
df[["Sales", "Profit", "Profit Margin"]].head()
```

```python
total_sales = df["Sales"].sum()
total_profit = df["Profit"].sum()
total_orders = df["Order ID"].nunique()
total_quantity = df["Quantity"].sum()
avg_order_value = df.groupby("Order ID")["Sales"].sum().mean()
avg_discount = df["Discount"].mean()
avg_profit_margin = df["Profit Margin"].mean()
```

```python
print("Total Sales:", round(total_sales, 2))
print("Total Profit:", round(total_profit, 2))
print("Total Orders:", total_orders)
print("Total Quantity:", total_quantity)
print("Average Order Value:", round(avg_order_value, 2))
print("Average Discount:", round(avg_discount, 2))
print("Average Profit Margin:", round(avg_profit_margin, 2))
```

```python
monthly_sales = df.groupby("Order Year-Month")["Sales"].sum().reset_index()
```

```python
monthly_sales.plot(x="Order Year-Month", y="Sales", kind="area", figsize=(10,3), color="skyblue")
plt.xticks(rotation=45)
plt.title("Monthly Sales Trend")
plt.show()
```

```python
monthly_profit = df.groupby("Order Year-Month")["Profit"].sum().reset_index()
```

```python
ax = monthly_profit.plot(x="Order Year-Month", y="Profit", kind="bar", figsize=(10,3), color="skyblue")
plt.xticks(range(0, len(monthly_profit), 3), monthly_profit["Order Year-Month"][::3], rotation=45)
plt.title("Monthly Profit Trend")
plt.show()
```

```python
category_sales_profit = df.groupby("Category")[["Sales", "Profit"]].sum().reset_index()
category_sales_profit
```

```python
category_sales_profit.plot(x="Category", y="Profit", kind="bar", figsize=(8,5), color="skyblue")
plt.title("Profit by Category")
plt.xticks(rotation=0)
plt.show()
```

```python
subcat_profit = df.groupby("Sub-Category")["Profit"].sum().reset_index().sort_values(by="Profit", ascending=False)
subcat_profit
```

```python
subcat_profit.plot(x="Sub-Category", y="Profit", kind="bar", figsize=(12,5), color="skyblue")
plt.title("Profit by Sub-Category")
plt.xticks(rotation=45)
plt.show()
```

```python
region_sales = df.groupby("Region")["Sales"].sum().reset_index().sort_values(by="Sales", ascending=False)
region_sales
```

```python
region_sales.plot(x="Region", y="Sales", kind="bar", figsize=(8,5), color="skyblue")
plt.title("Sales by Region")
plt.show()
```

```python
region_profit = df.groupby("Region")["Profit"].sum().reset_index().sort_values(by="Profit", ascending=False)
region_profit
```

```python
region_profit.plot(x="Region", y="Profit", kind="bar", figsize=(8,5), color="skyblue")
plt.title("Profit by Region")
plt.show()
```

```python
discount_profit = df.groupby("Discount")["Profit"].mean().reset_index()
discount_profit
```

```python
discount_profit.plot(x="Discount", y="Profit", kind="line", figsize=(8,5), color="skyblue", marker="o")
plt.title("Average Profit by Discount")
plt.show()
```

```python
segment_sales = df.groupby("Segment")["Sales"].sum().reset_index().sort_values(by="Sales", ascending=False).round(2)
segment_sales
```

```python
segment_sales.plot(x="Segment", y="Sales", kind="bar", figsize=(8,5), color="skyblue")
plt.title("Sales by Segment")
plt.xticks(rotation=0)
plt.show()
```

```python
segment_profit = df.groupby("Segment")["Profit"].sum().reset_index().sort_values(by="Profit", ascending=False).round(2)
segment_profit
```

```python
segment_profit.plot(x="Segment", y="Profit", kind="barh", figsize=(8,5), color="skyblue")
plt.title("Profit by Segment")
plt.xticks(rotation=0)
plt.show()
```

```python
region_profit = df.groupby("Region")["Profit"].sum()
region_profit = region_profit[region_profit > 0]

segment_profit = df.groupby("Segment")["Profit"].sum()
segment_profit = segment_profit[segment_profit > 0]

plt.figure(figsize=(14,7))

plt.subplot(1, 2, 1)
plt.pie(
    region_profit,
    labels=region_profit.index,
    autopct="%1.1f%%",
    startangle=90,
    wedgeprops={"width": 0.6}
)
plt.title("Profit Share by Region")

plt.subplot(1, 2, 2)
plt.pie(
    segment_profit,
    labels=segment_profit.index,
    autopct="%1.1f%%",
    startangle=90,
    wedgeprops={"width": 0.6}
)
plt.title("Profit Share by Segment")

plt.tight_layout()
plt.show()
```
